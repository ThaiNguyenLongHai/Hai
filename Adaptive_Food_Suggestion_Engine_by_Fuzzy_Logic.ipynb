```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl


Taste =ctrl.Antecedent(np.arange(0,11,1),'Taste')
Environment=ctrl.Antecedent(np.arange(0,11,1),'Environment')   
Budget=ctrl.Antecedent(np.arange(0,11,1),'Budget')
User_Review=ctrl.Antecedent(np.arange(0,11,1),'User_Review')
Location=ctrl.Antecedent(np.arange(0,11,1),'Location')
Level=ctrl.Consequent(np.arange(0,11,1),'Level') 

Taste['Low']=fuzz.trimf(Taste.universe,[0,0,5])
Taste['Medium']=fuzz.trimf(Taste.universe,[0,5,10])
Taste['High']=fuzz.trimf(Taste.universe,[5,10,10])

Environment['Low']=fuzz.trimf(Environment.universe,[0,0,5])
Environment['Medium']=fuzz.trimf(Environment.universe,[0,5,10])
Environment['High']=fuzz.trimf(Environment.universe,[5,10,10])

Budget['Less_Preferred']=fuzz.trimf(Budget.universe,[0,0,5])
Budget['Preferred']=fuzz.trimf(Budget.universe,[0,5,10])
Budget['More_Preferred']=fuzz.trimf(Budget.universe,[5,10,10])

User_Review['Less_Preferred']=fuzz.trimf(User_Review.universe,[0,0,5])
User_Review['Preferred']=fuzz.trimf(User_Review.universe,[0,5,10])
User_Review['More_Preferred']=fuzz.trimf(User_Review.universe,[5,10,10])

Location['Close']=fuzz.trimf(Location.universe,[0,0,5])
Location['Near']=fuzz.trimf(Location.universe,[0,5,10])
Location['Far']=fuzz.trimf(Location.universe,[5,10,10])

Level['Level 1']=fuzz.trimf(Level.universe,[0,2,4])
Level['Level 2']=fuzz.trimf(Level.universe,[2,4,6])
Level['Level 3']=fuzz.trimf(Level.universe,[4,6,8])
Level['Level 4']=fuzz.trimf(Level.universe,[6,8,10])
Level['Level 5']=fuzz.trimf(Level.universe,[8,10,10])

Taste.view()
Environment.view()
Budget.view()
User_Review.view()
Location.view()
Level.view()
```
